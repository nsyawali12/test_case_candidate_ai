```python
import pandas as pd
```

```python
# Define file path
sim_features_path = "../results/similarity_features.csv"
label_file_path = '../../Dataset/CodeAid_Source_Codes_Labeling.xlsx'
output_path = "../results/merge_features_and_label.csv"

# Load dataset
sim_features = pd.read_csv(sim_features_path)
label_file = pd.read_excel(label_file_path)
```
